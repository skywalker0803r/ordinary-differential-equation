```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def model(y,t):
  dydt = r*y*(1-y/K)
  return dydt
y0 = 3.8
r = 0.079
K = 200
t = np.linspace(0,100,10000)
y = odeint(model,y0,t)
plt.figure()
plt.plot(t,model(y,t),linewidth=3,color='red')
plt.legend(["dP/dt"])
plt.show()
```

```python
plt.plot(t,y,linewidth=3,color='red')
plt.legend(["P"])
plt.show()
```
